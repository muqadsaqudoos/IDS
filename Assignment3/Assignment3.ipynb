```python
def f2(orders):
    a =filter(lambda  x: isinstance(x["total"],(int,float)) and int(x["total"])>=0,orders)
    a = list(a)
    b = map(lambda x: {**x, "total": x["total"]*0.10+x["total"] if x["total"]>300 else x["total"]}, a)
    return list(b)

orders = [
{"customer": "Alice", "total": 250.5},
{"customer": "Bob", "total": "invalid_data"},
{"customer": "Charlie", "total": 450},
{"customer": "Daisy", "total": 100.0},
{"customer": "Eve", "total": -30},
]
a = f2(orders)
print(a)
```

```python
from functools import reduce
def f3(orders):

  a =filter(lambda  x: isinstance(x["total"],(int,float)) and int(x["total"])>=0,orders)
  a = list(a)
  b = map(lambda x: {**x, "total": x["total"]*0.10+x["total"] if x["total"]>300 else x["total"]}, a)
  b = list(b)
  c = reduce(lambda x,y: x+y['total'],b,0)
  return c

orders = [
{"customer": "Alice", "total": 250.5},
{"customer": "Bob", "total": "invalid_data"},
{"customer": "Charlie", "total": 450},
{"customer": "Daisy", "total": 100.0},
{"customer": "Eve", "total": -30},
]
a = f3(orders)
print(a)
```

```python
class SquareIterator:
    def __init__(self, n):
        self.n = n

    def __iter__(self):
        for i in range(1, self.n + 1):
            yield i ** 2

a = SquareIterator(10)
for i in a:
    print(i)
```

```python
def fibonacci_generator():
    n1, n2 = 0, 1
    while True:
        yield n1
        n1, n2 = n2, n1 + n2

fib = fibonacci_generator()
for i in range(10):
    print(next(fib))
```

```python
class DivisionByZeroError(Exception):
    pass

def divide_numbers(numbers, divisor):
    if divisor == 0:
        raise DivisionByZeroError("Cannot divide by zero.")

    result = []
    for num in numbers:
        try:
            result.append(num / divisor)
        except Exception as e:
            raise Exception(f"Error dividing '{num}'") from e

    return result

numbers = [10, 20, 'invalid', 40]

try:
    print(divide_numbers(numbers, 2))
except DivisionByZeroError as e:
    print(f"Custom error: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def exception_logger(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error in function '{func.__name__}': {type(e).__name__} - {e}")
            raise
    return wrapper

@exception_logger
def divide(x, y):
    return (x / y)

try:
    divide(10, 0)
except ZeroDivisionError:
    print("Cannot divide by zero. Please try again with different values.")
```
